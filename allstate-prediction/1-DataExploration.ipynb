### Now, are there any final options that are very common?

```python
def formatter(row):
    return ''.join([str(i) for i in row.values.ravel().tolist()])

cols = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
this_data = train_data_all[train_data_all['record_type'] == 1]
this_data = this_data[cols].apply(formatter, axis=1)

print(this_data.count())
freqs = this_data.groupby(this_data).count()
very_frequent = freqs[freqs > 1000]
print(very_frequent)
print(very_frequent.sum())
```

### Do we get more confident for a later shopping_pt?

```python
cols = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

def formatter(row):
    return ''.join([str(i) for i in row.values.ravel().tolist()])

def leads_getter(i):
    return leads[leads['shopping_pt']==i][cols].apply(formatter, axis=1)
    
buyer_results = buyers[cols].apply(formatter, axis=1)
```

```python
(buyer_results == leads_getter(10)).sum()
```

```python
def fn(col):
    return (buyers[col] == leads[leads['shopping_pt']==1][col]).sum() / len(buyer_results.index)
    
cols = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
res = [fn(col) for col in cols]

plt.figure()
plt.bar(range(len(cols)), res, width=0.8)
```

```python
'{0:.1%}'.format((buyer_results == leads_results).sum() / len(buyer_results.index))
```
