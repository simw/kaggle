### shopping_pt distribution

```python
buyers = train_data_all[train_data_all['record_type'] == 1]
leads = train_data_all
```

```python
b = pd.DataFrame()
l = pd.DataFrame()
b[['shopping_pt', 'buyers']] = buyers[['customer_ID','shopping_pt']].groupby('shopping_pt').count().reset_index()
l[['shopping_pt', 'leads']] = leads[['customer_ID','shopping_pt']].groupby('shopping_pt').count().reset_index()
ratio = pd.merge(b, l, on='shopping_pt', how='outer')
ratio = ratio.sort(['shopping_pt'])
ratio['buyers'] = ratio['buyers'].fillna(0)
ratio['ratio'] = ratio['buyers'] / ratio['leads']
ratio['cum_buyers'] = ratio['buyers'].cumsum()
ratio['cum_leads'] = ratio['leads'].cumsum()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))

ratio.plot(ax=ax1, kind='bar', x='shopping_pt', y='leads', alpha=0.35, label=['Shoppers'])
ratio.plot(ax=ax1, kind='bar', x='shopping_pt', y='buyers', alpha=0.65)
ax1.set_title('Shoppers and buyers for each shopping point')

ratio.plot(ax=ax2, kind='bar', x='shopping_pt', y='cum_leads', alpha=0.35, label=['Shoppers'])
ratio.plot(ax=ax2, kind='bar', x='shopping_pt', y='cum_buyers', alpha=0.65)
ax2.set_title('Shoppers and buyers for each shopping point')

ratio.plot(ax=ax3, kind='bar', x='shopping_pt', y='ratio', alpha=0.5, legend=False)
ax3.set_title('Proportion of shoppers that are buyers')
```

### Confidence by shopping_pt

```python
def add_will_change_col(df):
    conv = '1000000*A + 100000*B + 10000*C + 1000*D + 100*E + 10*F + 1*G'
    
    df.eval('current_plan = ' + conv)
    tmp = pd.DataFrame()
    tmp[['customer_ID', 'purchased_plan']] = df[df['record_type'] == 1][['customer_ID', 'current_plan']]
    df = pd.merge(df, tmp, on='customer_ID', how='left')
    df['will_change'] = (df['purchased_plan'] != df['current_plan'])
    return df
```

```python
shoppers = add_will_change_col(train_data_all)
no_changers = shoppers[shoppers['will_change'] == False]

nc = pd.DataFrame()
s = pd.DataFrame()
nc[['shopping_pt', 'no_change']] = no_changers[['customer_ID','shopping_pt']].groupby('shopping_pt').count().reset_index()
s[['shopping_pt', 'all']] = shoppers[['customer_ID','shopping_pt']].groupby('shopping_pt').count().reset_index()
ratio = pd.merge(nc, s, on='shopping_pt', how='outer')
ratio = ratio.sort(['shopping_pt'])
ratio['ratio'] = ratio['no_change'] / ratio['all']

ratio['cum_no_change'] = ratio['no_change'].cumsum()
ratio['cum_all'] = ratio['all'].cumsum()
ratio['cum_ratio'] = ratio['cum_no_change'] / ratio['cum_all']
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(24,6))

ratio.plot(ax=ax1, kind='bar', x='shopping_pt', y='all', alpha=0.35, label=['Shoppers'])
ratio.plot(ax=ax1, kind='bar', x='shopping_pt', y='no_change', alpha=0.65)
ax1.set_title('Shoppers and final decision for each shopping point')

ratio.plot(ax=ax2, kind='bar', x='shopping_pt', y='cum_all', alpha=0.35, label=['Shoppers'])
ratio.plot(ax=ax2, kind='bar', x='shopping_pt', y='cum_no_change', alpha=0.65)
ax2.set_title('Shoppers and final decision for each shopping point')

ratio.plot(ax=ax3, kind='bar', x='shopping_pt', y='ratio', alpha=0.5, legend=False)
ax3.set_title('Proportion of shoppers that will not change later')

ratio.plot(ax=ax4, kind='bar', x='shopping_pt', y='cum_ratio', alpha=0.5, legend=False)
ax4.set_title('Proportion of shoppers that will not change later')
```

### Explore the insurance options

```python
def add_will_change_col(df):
    conv = '1000000*A + 100000*B + 10000*C + 1000*D + 100*E + 10*F + 1*G'
    
    df.eval('current_plan = ' + conv)
    tmp = pd.DataFrame()
    tmp[['customer_ID', 'purchased_plan']] = df[df['record_type'] == 1][['customer_ID', 'current_plan']]
    df = pd.merge(df, tmp, on='customer_ID', how='left')
    df['will_change'] = (df['purchased_plan'] != df['current_plan'])
    return df

shoppers = add_will_change_col(train_data_all)
shoppers.columns
```

```python
f, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6))

shoppers['cost'].plot(ax=ax1, kind='hist', bins=50, alpha=0.5)
shoppers[shoppers['will_change']]['cost'].plot(ax=ax1, kind='hist', bins=50)

count_shoppers, division = np.histogram(shoppers['cost'], bins=50)
count_will_change = np.histogram(shoppers[shoppers['will_change']]['cost'], bins=division)[0]
count = pd.DataFrame()
count['cost'] = division[:-1]
count['shoppers'] = count_shoppers
count['will_change'] = count_will_change
count['ratio'] = count['will_change'] / count['shoppers']
count.plot(ax=ax2, x='cost', y='ratio', kind='line')
```

```python
cutoff = 885
print('Cost over {0}, proportion who change: {1:.4%}'.format(cutoff, (shoppers[shoppers['will_change']]['cost'] > cutoff).sum() / 
                      (shoppers['cost'] > cutoff).sum()))
print('Cost over {0} / total number of shoppers = {1:.4%}'.format(cutoff, (shoppers['cost'] > cutoff).sum() / shoppers['cost'].count()))
```
