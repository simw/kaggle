### Plot figures

```python
n_digits = 16
labels = train_data_all.loc[:n_digits-1, 'label']
digits = train_data_all.loc[:n_digits-1, 'pixel0':].values

n = np.sqrt(digits[0].size)
digits = np.reshape(digits, (n_digits, n, n))

f, axs = plt.subplots(4, 4, figsize=(10, 10))
for i in range(4):
    for j in range(4):
        n = i*4 + j
        axs[i, j].matshow(digits[n], cmap=plt.cm.gray)
        axs[i, j].set_title(labels[n])
```

### Look at the fourier transform of the images

```python
from scipy import fftpack
```

```python
n_digits = 5
labels = train_data_all.loc[:n_digits-1, 'label']
digits = train_data_all.loc[:n_digits-1, 'pixel0':].values

n = np.sqrt(digits[0].size)
digits = np.reshape(digits, (n_digits, n, n))

f, axs = plt.subplots(n_digits, 2, figsize=(8, 4*n_digits))

for i in range(n_digits):
    f1 = fftpack.fft2(digits[i])
    f2 = fftpack.fftshift(f1)
    psd2d = np.abs(f2)**2
    
    axs[i, 0].imshow(digits[i])
    axs[i, 1].imshow(psd2d)
```
