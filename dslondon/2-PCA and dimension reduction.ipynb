## Import data

```python
cwd = '/home/simon/Programming/data/sklearn/kaggle-dslondon'

train_data = pd.read_csv(cwd + '/data/train.csv', header=None)
train_target = pd.read_csv(cwd + '/data/trainLabels.csv', header=None)
test_data = pd.read_csv(cwd + '/data/test.csv', header=None)

print('Training rows = {0}'.format(train_data.count()[0]))
print('Target rows for training set = {0}'.format(train_target.count()[0]))
print('Test rows = {0}'.format(test_data.count()[0]))
```

## Try some PCA

```python
from sklearn.decomposition import PCA, RandomizedPCA

from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.cross_validation import train_test_split
```

```python
n = 20

pca = PCA(n_components=n)
proj = pca.fit_transform(train_data)
print('Total = {0:.1%}, components {1}'.format(
    pca.explained_variance_ratio_.sum(), pca.explained_variance_ratio_))
plt.scatter(proj[:, 0], proj[:, 1], c=train_target)
```

```python
def run_pca(n):
    pca = PCA(n_components=n)
    pca.fit(train_data)
    return pca.explained_variance_ratio_.sum()

xs = range(1, 40, 1)
ys = [run_pca(n) for n in xs]
plt.plot(xs, ys)
```

### Now run our classifiers on the reduced set of data

Samples of all classifiers

```python
names = ['Gaussian Naive Bayes', 'Decision Tree', 'Random Forest', 'K Nearest Neighbours', 'Support Vector Machines']
clfs = [
    GaussianNB(), 
    DecisionTreeClassifier(max_depth=5), 
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
    KNeighborsClassifier(3),
    SVC()]

x_train, x_test, y_train, y_test = train_test_split(proj, train_target.values.ravel())

for name, clf in zip(names, clfs):
    clf.fit(x_train, y_train)
    score = clf.score(x_test, y_test)

    print('{0} = {1:.1%}'.format(name, score))
```

SVC for different dimensions of PCA

```python
x_train, x_test, y_train, y_test = train_test_split(train_data.values, train_target.values.ravel())

def run_pca(n):
    pca = PCA(n_components=n)
    pca_fit = pca.fit(x_train)
    proj = pca_fit.transform(x_train)
    clf = SVC()
    clf.fit(proj, y_train)
    return clf.score(pca_fit.transform(x_test), y_test)

xs = range(1,41)
ys = [run_pca(n) for n in xs]
f, (axs1, axs2) = plt.subplots(1, 2)

axs1.plot(xs, ys)
axs2.plot(xs, ys)
axs2.set_ylim([0.85, 0.95])
```

k nearest neighbours for different dimensions of PCA

```python
x_train, x_test, y_train, y_test = train_test_split(train_data.values, train_target.values.ravel())

def run_pca(n):
    pca = PCA(n_components=n)
    pca_fit = pca.fit(x_train)
    proj = pca_fit.transform(x_train)
    clf = KNeighborsClassifier(6)
    clf.fit(proj, y_train)
    return clf.score(pca_fit.transform(x_test), y_test)

xs = range(1,41)
ys = [run_pca(n) for n in xs]
f, (axs1, axs2) = plt.subplots(1, 2)

axs1.plot(xs, ys)
axs2.plot(xs, ys)
axs2.set_ylim([0.85, 0.95])
```

## Try some Manifold

```python
from sklearn.manifold import Isomap, LocallyLinearEmbedding
```

```python
n = 15

iso = Isomap(n_neighbors=5*n, n_components=n)
proj = iso.fit_transform(train_data)
plt.scatter(proj[:, 0], proj[:, 1], c=train_target)

x_train, x_test, y_train, y_test = train_test_split(proj, train_target.values.ravel())
clf = SVC()
clf.fit(x_train, y_train)
print('Score = {0:.1%}'.format(clf.score(x_test, y_test)))
```

```python
n = 2

lle = LocallyLinearEmbedding(n_neighbors=5*n, n_components=n, method='modified')
proj = lle.fit_transform(train_data)
plt.scatter(proj[:, 0], proj[:, 1], c=train_target)

x_train, x_test, y_train, y_test = train_test_split(proj, train_target.values.ravel())
clf = SVC()
clf.fit(x_train, y_train)
print('Score = {0:.1%}'.format(clf.score(x_test, y_test)))
```
