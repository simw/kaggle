## Fiting a multi-dimensional, zero covariance gaussian to the data

```python
train0 = train_data[train_target[0] == 0]
train1 = train_data[train_target[0] == 1]

means0 = range(40)
stdevs0 = range(40)
means1 = range(40)
stdevs1 = range(40)

for n in range(40):
    means0[n] = train0[n].mean()
    stdevs0[n] = train0[n].std()
    means1[n] = train1[n].mean()
    stdevs1[n] = train1[n].std()
```

```python
def gaussian(x, mu, sigma):
    return 1 / np.sqrt(2*np.pi*np.power(sigma, 2)) * np.exp(-np.power(x-mu, 2) / (2.*np.power(sigma, 2)))

def clfer(s):
    p0 = 1
    p1 = 1
    for n in range(40):
        p0 *= gaussian(s[n], means0[n], stdevs0[n])
        p1 *= gaussian(s[n], means1[n], stdevs1[n])
        
    return int(p1 > p0)
```

```python
predictions = np.array([clfer(train_data.loc[i, :]) for i in range(1000)])
```

```python
(predictions == train_target.values.ravel()).sum()
```

## Fitting proper multi-dimensional gaussian

```python
from scipy.stats import multivariate_normal
```

Try just a subset of columns first

```python
cols = [6,12,14,18,32,34,36,39]
train_subdata = train_data[cols]

train0 = train_subdata[train_target[0] == 0]
train1 = train_subdata[train_target[0] == 1]

means0 = train0.mean()
cov0 = train0.cov()

means1 = train1.mean()
cov1 = train1.cov()

var0 = multivariate_normal(means0, cov0)
var1 = multivariate_normal(means1, cov1)

def clfer(s):
    p0 = var0.pdf(s[cols])
    p1 = var1.pdf(s[cols])
        
    return int(p1 > p0)

predictions = np.array([clfer(train_data.loc[i, :]) for i in range(1000)])
(predictions == train_target.values.ravel()).sum()
```

```python
train0 = train_data[train_target[0] == 0]
train1 = train_data[train_target[0] == 1]

means0 = train0.mean()
cov0 = train0.cov()

means1 = train1.mean()
cov1 = train1.cov()

var0 = multivariate_normal(means0, cov0)
var1 = multivariate_normal(means1, cov1)

def clfer(s):
    p0 = var0.pdf(s)
    p1 = var1.pdf(s)
        
    return int(p1 > p0)

predictions = np.array([clfer(train_data.loc[i, :]) for i in range(1000)])
(predictions == train_target.values.ravel()).sum()
```
